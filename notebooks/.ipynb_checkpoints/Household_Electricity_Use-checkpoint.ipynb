Source: 
https://archive.ics.uci.edu/ml/datasets/Individual+household+electric+power+consumption

**Data Set Information:**

This archive contains 2.075.259 measurements gathered in a house located in Sceaux (7km of Paris, France) between December 2006 and November 2010 (47 months).  

Notes: 
1. (global_active_power*1000/60 - sub_metering_1 - sub_metering_2 - sub_metering_3) represents the active energy consumed every minute (in watt hour) in the household by electrical equipment not measured in sub-meterings 1, 2 and 3. 
2. The dataset contains some missing values in the measurements (nearly 1,25% of the rows). All calendar timestamps are present in the dataset but for some timestamps, the measurement values are missing: a missing value is represented by the absence of value between two consecutive semi-colon attribute separators. For instance, the dataset shows missing values on April 28, 2007.


**Attribute Information:**

1. date: Date in format dd/mm/yyyy    
2. time: time in format hh/:mm/:ss    
3. global_active_power: household global minute-averaged active power (in kilowatt)    
4. global_reactive_power: household global minute-averaged reactive power (in kilowatt)    
5. voltage: minute-averaged voltage (in volt)    
6. global_intensity: household global minute-averaged current intensity (in ampere)    
7. sub_metering_1: energy sub-metering No. 1 (in watt-hour of active energy). It corresponds to the kitchen, containing mainly a dishwasher, an oven and a microwave (hot plates are not electric but gas powered).    
8. sub_metering_2: energy sub-metering No. 2 (in watt-hour of active energy). It corresponds to the laundry room, containing a washing-machine, a tumble-drier, a refrigerator and a light.    
9. sub_metering_3: energy sub-metering No. 3 (in watt-hour of active energy). It corresponds to an electric water-heater and an air-conditioner.

## 2.3 Reshape input and ground truth data for feeding into RNN
[Back to contents](#top)

```python
# Does using sequences of 16/other make sense?
# - perhaps human expert needs longer sequences to be able to label the dataset
# - but shorter sequences means the model can be better at predicting closer to real-time
# - Tx=16 already means a segment of 80 mins
start = 1115
length = 16

f, axarr = plt.subplots(3, sharex=True, figsize=(8,6))
sns.lineplot(x=samples_35.iloc[start:start+length, :].index, y='s1', data=samples_35.iloc[start:start+length, :], hue='operational_mode', ax=axarr[0])
sns.lineplot(x=samples_35.iloc[start:start+length, :].index, y='s2', data=samples_35.iloc[start:start+length, :], hue='operational_mode', ax=axarr[1])
sns.lineplot(x=samples_35.iloc[start:start+length, :].index, y='s3', data=samples_35.iloc[start:start+length, :], hue='operational_mode', ax=axarr[2])
plt.show()
```

```python
Tx = 16             # In this case, Ty = Tx

n_sensors  = 3
n_op_modes = df['operational_mode'].unique().shape[0]
n_assets   = max(df['asset_id'])

m_seq = 0
for i in range(n_assets):
    df_temp = df[df['asset_id'] == i+1]
    m_seq += len(df_temp)//Tx

print('Total number of sequence examples is', m_seq, 'when grouped by', Tx, 'elements for each sequence.')
print('Loss of', len(df) - m_seq*Tx, 'lines of data. May be corrected later by inclusion with padding.')
```

```python
# Map operational mode to integer labels
mode_label_dict = {'a': 0, 'b': 1, 'c': 2, 'd': 3, 'e': 4, 'f': 5}

df['op_mode_label'] = df['operational_mode'].map(mode_label_dict)
df.head()
```

```python
# Want standardised sensor data X to be in form (m_seq, Tx, n_sensors)
# ............                 and y be in form (m_seq, Tx, n_op_modes)
X = np.zeros((m_seq, Tx, n_sensors))
y = np.zeros((m_seq, Tx, 1))  # Use one-hot encoding later

asset_start = 0 
for asset in range(n_assets):
    df_asset = df[df['asset_id']==asset+1].sort_index() 
    n_asset_examples = len(df_asset)//Tx
    for i in range(n_asset_examples):
        for j in range(Tx):
            for k in range(n_sensors):
                X[asset_start + i, j, k] = df_asset.iloc[i*Tx+j, k+5]
                
            y[asset_start + i, j, :] = df_asset.iloc[i*Tx+j, 8]
            
    asset_start += n_asset_examples
```

```python
print('The shape of X is', X.shape)
print('The shape of y is', y.shape)
```

As an initial trial, we will consider the Gated Recurrent Unit (GRU) model, with the following architecture:
<img src="./model_sketch.jpg" style="width: 500px;"/>

```python
from keras.models import Model, load_model
from keras.layers import Dense, Input, Dropout, GRU, LSTM, Activation, TimeDistributed, BatchNormalization
from keras.preprocessing import sequence
from keras.initializers import glorot_uniform
from keras import backend as K
np.random.seed(5)
```

```python
K.clear_session()

def gru_model(input_shape):

    # Define input to model
    time_seq = Input(shape = input_shape, dtype='float32')
    
    # Create layers
    X = GRU(128, activation='tanh', return_sequences=True, name='gru_1')(time_seq)
    X = Dropout(0.8, name='dropout_1')(X)
    
    X = GRU(128, activation='tanh', return_sequences=True, name='gru_2')(X)
    X = Dropout(0.8, name='dropout_2')(X)
    
    X = TimeDistributed(Dense(6, activation='softmax'), name='time_distributed_softmax')(X)
        
    # Create Model instance which converts input sequence into X.
    model = Model(inputs=time_seq, outputs=X)
    
    return model
```

```python
model = gru_model(input_shape = (Tx, n_sensors))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy' , optimizer='adam', metrics=['categorical_accuracy'])
```

```python
# Load saved model
model = load_model('gru_2layer_model_200epochs.h5')
```

```python
gru_history = model.fit(train_X, train_y, epochs=10, batch_size=32, shuffle=True)
```

```python
# loss, acc = model.evaluate(train_X, train_y)
# print("Train set accuracy = ", acc)

loss, acc = model.evaluate(dev_X, dev_y)
print("Dev set accuracy = ", acc)
```

## 4.1 Examples of predicted versus actual operational modes
[Back to contents](#top)

```python
def make_prediction(model, x_input, n_op_modes):
    """ 
    Function that uses the model to make one-hot predictions given the standardised sensor readings 
    Outputs one-hot predictions of shape (number_of_sequences, Tx, n_op_modes)
    """
    output     = model.predict(x_input)
    prediction = np.eye(n_op_modes)[np.argmax(output[:,:,:], axis=2)]
    
    return prediction

def one_hot_to_categories(one_hot_vectors):
    """
    Function that converts operational mode one-hot vectors into original labels
    Outputs predictions of shape (number_of_sequences, Tx, 1)
    """
    labels          = np.argmax(one_hot_vectors, axis=2)  
    label_mode_dict = {0: 'a', 1: 'b', 2: 'c', 3: 'd', 4: 'e', 5: 'f'}
    labels          = np.vectorize(label_mode_dict.get)(labels)

    return labels

def devset_crude_compare_predictions_actuals(start_loc, num_of_seqs=1):
    """
    An initial crude method to visualise predictions vs. actuals.
    Prints out the predictions and ground-truths of the start_loc-th sequence in the dev set,
    and the subsequent num_of_seq - 1 sequences. 
    """
    example       = [start_loc+i for i in range(num_of_seqs)]
    one_hot_pred  = make_prediction(model, dev_X[example,:,:], n_op_modes)
    predictions   = one_hot_to_categories(one_hot_pred)
    actuals       = one_hot_to_categories(dev_y[example])

    print('Predictions:')
    print(predictions)
    print()
    print('Actuals:')
    print(actuals)
    
    return
```

```python
# Choose a example sequences in dev set and make prediction - crude way to compare for now...
devset_crude_compare_predictions_actuals(start_loc=3000, num_of_seqs=5)
```

<a id='confusion_matrix'></a>

## 4.2 Confusion matrix
[Back to contents](#top)

```python
from sklearn.metrics import confusion_matrix

y_true = np.ravel( one_hot_to_categories(dev_y) )

y_pred = make_prediction(model, dev_X, n_op_modes)
y_pred = np.ravel( one_hot_to_categories(y_pred) )

labels = ['a', 'b', 'c', 'd', 'e', 'f']
conf_matrix = confusion_matrix(y_true, y_pred, labels=labels)
```

```python
conf_matrix_df = pd.DataFrame(conf_matrix/np.sum(conf_matrix), index=labels, columns=labels)

plt.figure(figsize=(10,6))
sns.heatmap(conf_matrix_df, annot=True, fmt='.2%', linewidths=0.5)
plt.ylabel('Actual Operational Mode')
plt.xlabel('Predicted Operational Mode')
plt.title('Confusion Matrix for Dev Set')
```

## 4.3 Precision & Recall
[Back to contents](#top)

```python
# Vector of true positives
tp = np.zeros(n_op_modes)
for i in range(n_op_modes):
    tp[i] = conf_matrix[i][i]

# Vector of (true positives + false positives)
tp_fp = np.sum(conf_matrix, axis=0)

# Vector of (true positives + false negatives)
tp_fn = np.sum(conf_matrix, axis=1)
```

```python
recall = tp/tp_fn
precision = tp/tp_fp

label_mode_dict = {0: 'a', 1: 'b', 2: 'c', 3: 'd', 4: 'e', 5: 'f'}
for i in range(n_op_modes):
    print("Recall of '%s': %3.2f, Precision of '%s': %3.2f" % (label_mode_dict[i], recall[i], label_mode_dict[i], precision[i]))
```

__Notes:__
* Precision and recall values for operational modes 'a', 'b', 'e', and 'f' seem ok, depending on the performance obtainable by a human expert
* Precision and recall for 'c' and 'd' are poor
* Of the actual 'c' and 'd' instances, many are mis-predicted as 'b'
* Of the predicted 'c' instances many are actually 'd', and vice versa
